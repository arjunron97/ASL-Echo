```python
def detect_outliers(df, name="dataset", method='iqr', class_column='class',
                   output_plots=False, save_outliers=False,
                   ignore_zero_columns=True, z_threshold=3, iqr_multiplier=1.5):
    """
    Detect outliers in landmark coordinate data from a DataFrame.

    Parameters:
    - df (DataFrame): Input DataFrame containing the data
    - name (str): Name of the dataset (for reporting)
    - method (str): 'iqr', 'zscore', or 'range' (default: 'iqr')
    - class_column (str): Column name for grouping (default: 'class')
    - output_plots (bool): Whether to generate visualizations (default: False)
    - save_outliers (bool): Whether to save outliers to CSV (default: False)
    - ignore_zero_columns (bool): Whether to ignore all-zero columns (default: True)
    - z_threshold (float): Z-score threshold (default: 3)
    - iqr_multiplier (float): IQR multiplier (default: 1.5)

    Returns:
    - dict: Dictionary containing outliers DataFrame and summary statistics
    """

    # Identify coordinate columns
    coordinate_cols = [col for col in df.columns if col.endswith(('_x', '_y', '_z'))]

    # Ignore all-zero columns if requested
    if ignore_zero_columns:
        non_zero_cols = df[coordinate_cols].columns[(df[coordinate_cols] != 0).any()]
        coordinate_cols = [col for col in coordinate_cols if col in non_zero_cols]

    results = {
        'dataset_name': name,
        'total_rows': len(df),
        'coordinate_columns': coordinate_cols,
        'outliers_detected': None,
        'outliers_by_class': {}
    }

    # Method 1: Range-based detection
    if method == 'range':
        outliers = df[(df[coordinate_cols] < 0) | (df[coordinate_cols] > 1)]
        results['method'] = 'Range (0-1)'

    # Method 2: Z-score detection
    elif method == 'zscore':
        z_scores = df[coordinate_cols].apply(zscore)
        outliers = df[(z_scores.abs() > z_threshold).any(axis=1)]
        results['method'] = f'Z-score (threshold={z_threshold})'

    # Method 3: IQR detection (default)
    else:
        Q1 = df[coordinate_cols].quantile(0.25)
        Q3 = df[coordinate_cols].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_multiplier * IQR
        upper_bound = Q3 + iqr_multiplier * IQR
        outliers = df[(df[coordinate_cols] < lower_bound) | (df[coordinate_cols] > upper_bound)].dropna()
        results['method'] = f'IQR (multiplier={iqr_multiplier})'

    results['outliers_detected'] = len(outliers)
    results['outliers_percentage'] = (len(outliers) / len(df)) * 100

    # Group outliers by class if class column exists
    if class_column in df.columns:
        grouped = df.groupby(class_column)
        for group_name, group in grouped:
            if method == 'zscore':
                z_scores = group[coordinate_cols].apply(zscore)
                class_outliers = group[(z_scores.abs() > z_threshold).any(axis=1)]
            else:
                Q1 = group[coordinate_cols].quantile(0.25)
                Q3 = group[coordinate_cols].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - iqr_multiplier * IQR
                upper_bound = Q3 + iqr_multiplier * IQR
                class_outliers = group[(group[coordinate_cols] < lower_bound) |
                                      (group[coordinate_cols] > upper_bound)].dropna()

            results['outliers_by_class'][group_name] = {
                'count': len(class_outliers),
                'percentage': (len(class_outliers) / len(group)) * 100,
                'examples': class_outliers.head(3).to_dict('records')
            }

    # Generate visualizations
    if output_plots and coordinate_cols:
        plt.figure(figsize=(12, 6))

        # Boxplot for first 3 coordinate columns
        sns.boxplot(data=df[coordinate_cols[:3]])
        plt.title(f'Distribution of Coordinates ({name})\n({results["method"]})')
        plt.xticks(rotation=45)
        plt.show()

        # Scatter plot for first x-y pair if available
        if len(coordinate_cols) >= 2:
            x_col = next((col for col in coordinate_cols if col.endswith('_x')), None)
            y_col = next((col for col in coordinate_cols if col.endswith('_y')), None)
            if x_col and y_col:
                plt.figure(figsize=(10, 8))
                sns.scatterplot(data=df, x=x_col, y=y_col,
                               hue=class_column if class_column in df.columns else None,
                               alpha=0.6)
                plt.title(f'{name}: Coordinate Space - {x_col} vs {y_col}')
                plt.show()

    # Save outliers to file
    if save_outliers and len(outliers) > 0:
        output_file = f"{name}_outliers_{method}.csv"
        outliers.to_csv(output_file, index=False)
        results['output_file'] = output_file

    return {'results': results, 'outliers_df': outliers}
```

```python
# Example usage with multiple DataFrames
datasets = {
    "train": train_df,
    "val": val_df,
    "test": test_df
}

all_results = {}

for name, df in datasets.items():
    print(f"\nAnalyzing {name} dataset...")
    result = detect_outliers(
        df=df,
        name=name,
        method='iqr',
        class_column='class',
        output_plots=True,
        save_outliers=True
    )
    all_results[name] = result['results']

    # Print summary
    print(f"Found {result['results']['outliers_detected']} outliers ({result['results']['outliers_percentage']:.2f}%)")

    # Optionally inspect outliers DataFrame
    # display(result['outliers_df'].head())

# Print comprehensive results
print("\nFinal Summary:")
for name, res in all_results.items():
    print(f"\n{name.upper()} Dataset:")
    print(f"- Method: {res['method']}")
    print(f"- Total rows: {res['total_rows']}")
    print(f"- Outliers detected: {res['outliers_detected']} ({res['outliers_percentage']:.2f}%)")

    if res['outliers_by_class']:
        print("\nOutliers by class:")
        for cls, cls_data in res['outliers_by_class'].items():
            print(f"  {cls}: {cls_data['count']} ({cls_data['percentage']:.2f}%)")
```

```python
def detect_outliers_all_columns(df, name="dataset", method='iqr', class_column='class',
                               output_plots=False, save_outliers=False,
                               ignore_zero_columns=True, z_threshold=3, iqr_multiplier=1.5,
                               exclude_columns=None):
    """
    Detect outliers in all numeric columns of a DataFrame.

    Parameters:
    - df (DataFrame): Input DataFrame containing the data
    - name (str): Name of the dataset (for reporting)
    - method (str): 'iqr', 'zscore', or 'range' (default: 'iqr')
    - class_column (str): Column name for grouping (default: 'class')
    - output_plots (bool): Whether to generate visualizations (default: False)
    - save_outliers (bool): Whether to save outliers to CSV (default: False)
    - ignore_zero_columns (bool): Whether to ignore all-zero columns (default: True)
    - z_threshold (float): Z-score threshold (default: 3)
    - iqr_multiplier (float): IQR multiplier (default: 1.5)
    - exclude_columns (list): Columns to exclude from analysis (default: None)

    Returns:
    - dict: Dictionary containing outliers DataFrame and summary statistics
    """

    # Identify numeric columns and exclude specified columns
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    if exclude_columns:
        numeric_cols = [col for col in numeric_cols if col not in exclude_columns]
    if class_column in numeric_cols:
        numeric_cols.remove(class_column)

    # Ignore all-zero columns if requested
    if ignore_zero_columns:
        non_zero_cols = df[numeric_cols].columns[(df[numeric_cols] != 0).any()]
        numeric_cols = [col for col in numeric_cols if col in non_zero_cols]

    results = {
        'dataset_name': name,
        'total_rows': len(df),
        'analyzed_columns': numeric_cols,
        'outliers_detected': None,
        'outliers_by_class': {}
    }

    # Method 1: Range-based detection
    if method == 'range':
        outliers = df[(df[numeric_cols] < 0) | (df[numeric_cols] > 1)]
        results['method'] = 'Range (0-1)'

    # Method 2: Z-score detection
    elif method == 'zscore':
        z_scores = df[numeric_cols].apply(zscore)
        outliers = df[(z_scores.abs() > z_threshold).any(axis=1)]
        results['method'] = f'Z-score (threshold={z_threshold})'

    # Method 3: IQR detection (default)
    else:
        Q1 = df[numeric_cols].quantile(0.25)
        Q3 = df[numeric_cols].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_multiplier * IQR
        upper_bound = Q3 + iqr_multiplier * IQR
        outliers = df[((df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)).any(axis=1)]
        results['method'] = f'IQR (multiplier={iqr_multiplier})'

    results['outliers_detected'] = len(outliers)
    results['outliers_percentage'] = (len(outliers) / len(df)) * 100

    # Group outliers by class if class column exists
    if class_column in df.columns:
        grouped = df.groupby(class_column)
        for group_name, group in grouped:
            if method == 'zscore':
                z_scores = group[numeric_cols].apply(zscore)
                class_outliers = group[(z_scores.abs() > z_threshold).any(axis=1)]
            else:
                Q1 = group[numeric_cols].quantile(0.25)
                Q3 = group[numeric_cols].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - iqr_multiplier * IQR
                upper_bound = Q3 + iqr_multiplier * IQR
                class_outliers = group[((group[numeric_cols] < lower_bound) |
                                      (group[numeric_cols] > upper_bound)).any(axis=1)]

            results['outliers_by_class'][group_name] = {
                'count': len(class_outliers),
                'percentage': (len(class_outliers) / len(group)) * 100,
                'example_indices': class_outliers.head(3).index.tolist()
            }

    # Generate visualizations
    if output_plots and numeric_cols:
        # Boxplot grid for first 6 columns
        plt.figure(figsize=(16, 8))
        for i, col in enumerate(numeric_cols[:6], 1):
            plt.subplot(2, 3, i)
            sns.boxplot(x=df[col])
            plt.title(f'{col}\n({results["method"]})')
            plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Pairplot for first 3 numeric columns
        if len(numeric_cols) >= 3:
            sns.pairplot(df[numeric_cols[:3] + [class_column] if class_column in df.columns else numeric_cols[:3]],
                        hue=class_column if class_column in df.columns else None,
                        diag_kind='kde')
            plt.suptitle(f'{name}: Pairwise Relationships', y=1.02)
            plt.show()

    # Save outliers to file
    if save_outliers and len(outliers) > 0:
        output_file = f"{name}_outliers_{method}.csv"
        outliers.to_csv(output_file, index=False)
        results['output_file'] = output_file

    return {'results': results, 'outliers_df': outliers}
```

```python
def top_video_frames_by_class(df, top_n=100):
    """
    Finds the top N video IDs with the highest frame counts for each class.

    Args:
        df: DataFrame with 'video_id', 'frame', and 'class' columns.
        top_n: The number of top video IDs to retrieve for each class.

    Returns:
        A dictionary where keys are classes and values are lists of tuples,
        each tuple containing a video ID and its frame count.
    """
    top_videos = {}
    for class_name in df['class'].unique():
        class_df = df[df['class'] == class_name]
        frame_counts = class_df.groupby('video_id')['frame'].count().sort_values(ascending=False)
        top_videos[class_name] = list(frame_counts.head(top_n).items())  # List of (video_id, frame_count) tuples
    return top_videos


top_videos_dict = top_video_frames_by_class(train_df)

# Print the results nicely
for class_name, video_counts in top_videos_dict.items():
    print(f"Class: {class_name}")
    for video_id, frame_count in video_counts:
        print(f"  Video ID: {video_id}, Frame Count: {frame_count}")
    print("-" * 20)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import models, optimizers
from tensorflow.keras.layers import (
    Layer, Conv2D, BatchNormalization, Activation,
    GlobalAveragePooling2D, Dense, Dropout, Input, Add, GaussianNoise
)
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
import math

# ─── Utility: Normalize Adjacency Matrix ───────────────────────────────────────
def normalize_adjacency(A):
    I = np.eye(len(A))
    A_hat = A + I
    D = np.sum(A_hat, axis=1)
    D_inv_sqrt = np.diag(1.0 / np.sqrt(D))
    return D_inv_sqrt @ A_hat @ D_inv_sqrt

# ─── ST-GCN Block w/ Residual ──────────────────────────────────────────────────
class STGCNBlock(Layer):
    def __init__(self, out_channels, adjacency, t_kernel_size=9, stride=1, dropout_rate=0.3, **kwargs):
        super().__init__(**kwargs)
        self.out_channels = out_channels
        self.adjacency = np.array(adjacency)
        self.A_norm = tf.constant(normalize_adjacency(self.adjacency), dtype=tf.float32)
        self.t_kernel_size = t_kernel_size
        self.stride = stride
        self.conv_s = Conv2D(filters=out_channels, kernel_size=(1,1), padding='valid')
        self.conv_t = Conv2D(filters=out_channels,
                             kernel_size=(t_kernel_size,1),
                             strides=(stride,1),
                             padding='same')
        self.bn = BatchNormalization()
        self.act = Activation('relu')
        self.dropout = Dropout(dropout_rate)
        self.skip_conv = None

    def build(self, input_shape):
        batch, T, N, C = input_shape
        # build spatial and temporal convs
        self.conv_s.build(input_shape)
        shape_s = self.conv_s.compute_output_shape(input_shape)
        self.conv_t.build(shape_s)
        shape_t = self.conv_t.compute_output_shape(shape_s)
        self.bn.build(shape_t)
        # build skip connection if channel mismatch or stride !=1
        if C != self.out_channels or self.stride != 1:
            self.skip_conv = Conv2D(filters=self.out_channels, kernel_size=(1,1),
                                    strides=(self.stride,1), padding='same')
            self.skip_conv.build(input_shape)
        super().build(input_shape)

    def call(self, x, training=False):
        b, T, N, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        # graph conv
        x_ = tf.reshape(x, [b * T, N, C])
        x_ = tf.matmul(self.A_norm, x_)
        x_ = tf.reshape(x_, [b, T, N, C])
        # spatial conv
        x_ = self.conv_s(x_)
        # temporal conv
        x_ = self.conv_t(x_)
        x_ = self.bn(x_, training=training)
        # skip
        skip = x
        if self.skip_conv:
            skip = self.skip_conv(skip)
        # combine
        x_ = Add()([x_, skip])
        x_ = self.act(x_)
        return self.dropout(x_, training=training)

    def compute_output_shape(self, input_shape):
        batch, T, N, C = input_shape
        out_T = math.ceil(T / self.stride)
        return (batch, out_T, N, self.out_channels)

    def get_config(self):
        config = super().get_config()
        config.update({
            'out_channels': self.out_channels,
            'adjacency': self.adjacency.tolist(),
            't_kernel_size': self.t_kernel_size,
            'stride': self.stride
        })
        return config

# ─── Data Utilities ─────────────────────────────────────────────────────────────
def create_seq_lbl(df, feature_cols):
    """
    Extract sequences (list of arrays) and labels from DataFrame grouped by video_id.
    Each sequence is a 2D array (T, F).
    """
    seqs, labels = [], []
    for vid, grp in df.groupby('video_id'):
        seqs.append(grp[feature_cols].values)
        labels.append(grp['label'].iloc[0])
    return seqs, labels


def prepare_sequences(train_flat, val_flat, num_nodes, coord_dim, seq_len, pad_val):
    """
    Reshape flat sequences to (T, N, C), pad/truncate to seq_len, and stack into arrays.
    Returns X_train, X_val of shape (samples, seq_len, N, C).
    """
    def reshape_and_pad(flat_list):
        out = []
        for s in flat_list:
            # reshape to (T, N, C)
            s = s.reshape(s.shape[0], num_nodes, coord_dim)
            # truncate or pad
            if s.shape[0] >= seq_len:
                s2 = s[:seq_len]
            else:
                pad = np.full((seq_len - s.shape[0], num_nodes, coord_dim), pad_val, dtype=np.float32)
                s2 = np.concatenate([s, pad], axis=0)
            out.append(s2)
        return np.stack(out)
    X_train = reshape_and_pad(train_flat)
    X_val = reshape_and_pad(val_flat)
    return X_train, X_val


def get_skeleton_adjacency(num_nodes):
    """
    Create a simple chain adjacency or define your own graph.
    """
    A = np.zeros((num_nodes, num_nodes))
    # simple chain: each node connected to next
    for i in range(num_nodes - 1):
        A[i, i+1] = 1
        A[i+1, i] = 1
    return A

# ─── Build ST-GCN Model ─────────────────────────────────────────────────────────
def build_st_gcn(seq_len, num_nodes, coord_dim, adjacency, num_classes):
    inputs = Input(shape=(seq_len, num_nodes, coord_dim))
    x = GaussianNoise(0.01)(inputs)
    x = Conv2D(64, (1,1), padding='valid')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = STGCNBlock(64, adjacency, stride=1, dropout_rate=0.4)(x)
    x = STGCNBlock(128, adjacency, stride=2, dropout_rate=0.4)(x)
    x = STGCNBlock(256, adjacency, stride=2, dropout_rate=0.4)(x)
    x = STGCNBlock(256, adjacency, stride=2, dropout_rate=0.4)(x)
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    model.compile(
        optimizer=optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# ─── Main Execution ─────────────────────────────────────────────────────────────
if __name__ == '__main__':
    # Prepare feature columns
    cols = [c for c in train_scaled_df.columns if c not in ['video_id', 'label']]
    F = len(cols)
    coord_dim = 3 if F % 3 == 0 else 2
    num_nodes = F // coord_dim
    print(f"Features: {F}, CoordDim: {coord_dim}, Nodes: {num_nodes}")

    # Extract sequences and labels
    train_flat, train_lbls = create_seq_lbl(train_scaled_df, cols)
    val_flat, val_lbls = create_seq_lbl(val_scaled_df, cols)

    # Prepare data arrays
    SEQ_LEN, PAD = 30, -1.0
    X_train, X_val = prepare_sequences(train_flat, val_flat, num_nodes, coord_dim, SEQ_LEN, PAD)
    print(f"X_train: {X_train.shape}, X_val: {X_val.shape}")

    # Encode labels
    le = LabelEncoder()
    y_tr = to_categorical(le.fit_transform(train_lbls))
    y_val = to_categorical(le.transform(val_lbls))
    num_classes = y_tr.shape[1]
    print(f"Num classes: {num_classes}")

    # Build adjacency
    adjacency = get_skeleton_adjacency(num_nodes)

    # Callbacks
    ckpt = ModelCheckpoint('best_stgcn.weights.h5', save_best_only=True, monitor='val_loss')
    early = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

    # Build & train
    model = build_st_gcn(SEQ_LEN, num_nodes, coord_dim, adjacency, num_classes)
    model.fit(
        X_train, y_tr,
        validation_data=(X_val, y_val),
        epochs=150
        batch_size=8,
        callbacks=[ckpt],
        verbose=1
    )

    # Evaluate & Save
    loss, acc = model.evaluate(X_val, y_val, verbose=0)
    print(f"Val loss: {loss:.3f}, val_acc: {acc:.3f}")
    model.save('stgcn_model.keras')
```
