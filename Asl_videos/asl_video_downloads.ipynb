```python
#asl_train_videos
# Check if yt-dlp is installed
def find_yt_dlp():
    yt_dlp_path = shutil.which("yt-dlp")
    if yt_dlp_path:
        print(f"yt-dlp found at: {yt_dlp_path}")
        return yt_dlp_path
    else:
        raise FileNotFoundError("yt-dlp not found. Please ensure it is installed and in your PATH.")

# Function to download video using yt-dlp
def download_youtube_video(link, link_id, save_dir):
    try:
        yt_dlp_path = find_yt_dlp()

        # Construct the output filename using link_id
        output_filename = os.path.join(save_dir, f"{link_id}.%(ext)s")  

        # Use yt-dlp directly with subprocess
        cmd = [
            yt_dlp_path,
            '-o', output_filename,
            '--ignore-errors',
            '--no-playlist',
            '--add-metadata',
            '-f', 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
            link
        ]
        print(f"Running command: {' '.join(cmd)}")
        subprocess.run(cmd, check=True)
        print(f"Video downloaded successfully to {save_dir} with filename: {link_id}.mp4")
    except subprocess.CalledProcessError as e:
        print(f"Error downloading video from {link} with link_id {link_id}: {e}")
    except FileNotFoundError as e:
        print(e)

# Create a directory to store the downloaded videos
save_dir = r"D:\UNT\Research\ASL Echo\ASL-Echo\Asl_videos\asl_train_videos"
os.makedirs(save_dir, exist_ok=True)

# Load the dataset
asl_train = pd.read_json(r'D:\UNT\Research\ASL Echo\ASL-Echo\MSASL_transformed.json')

# Iterate through the DataFrame and download videos
for index, row in asl_train.iterrows():
    link = row.get('link')  # Using .get() to avoid KeyError
    link_id = row.get('link_id')

    if link and link_id:
        download_youtube_video(link, link_id, save_dir)
    else:
        print(f"Missing data at row {index}")

print("Video downloads complete.")
```
