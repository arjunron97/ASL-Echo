```python
import pyttsx3

engine = pyttsx3.init()

speaker = input("Gender voice:")
voices = engine.getProperty('voices')

if speaker == 'Male':
    engine.setProperty('voice', voices[0].id)
elif speaker == 'Female':
    engine.setProperty('voice', voices[1].id)
else:
    engine.setProperty('voice', voices[0].id)

words = ["Hi", "my", "name", "is", "arjun"]
sentence = ' '.join(words)


print(f"Speaking: {sentence}")
engine.say(sentence)
engine.runAndWait()


```

```python
# Change speaking rate
engine.setProperty('rate', 200)  # Default is around 200

# Change volume (0.0 to 1.0)
engine.setProperty('volume', 0.9)

# List available voices and choose one
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # voices[1] for a different voice
```
